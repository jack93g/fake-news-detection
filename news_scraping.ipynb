```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import ElementClickInterceptedException
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

pd.set_option('display.max_rows', 100)
```

```python
# function to scrape the text of the url from the href tag from the all sides website

def allsides_url(sites):
  
  lst = []
      
  for site in sites:

    url1 = f'https://www.allsides.com/{site}'

    response = requests.get(url1)
    html = response.content

    soup = BeautifulSoup(html, "html.parser")

    for tag in soup.findAll('div', class_ ='news-title'):
      for href in tag.findAll('a', href= True):
        lst.append(str(href['href']))

  return lst
```

```python
# get all topics urls from the topics page


url1 = 'https://www.allsides.com/topics-issues'

urls = []

response = requests.get(url1)
html = response.content

soup = BeautifulSoup(html, "html.parser")

for tag in soup.findAll('div', id="block-views-topic-menu-list-block-3", class_="block block-views first last odd"):
    for href in tag.findAll('a', href= True):
        urls.append(str(href['href']))

len(urls)
```

```python
urls_lst = allsides_url(urls)
len(urls_lst)
```

```python
# scrape headlines from all sides

def allsides_headlines(sites):
    
    urls = []

    for site in sites:

        url1 = f'https://www.allsides.com/{site}'


        response = requests.get(url1)
        html = response.content

        soup = BeautifulSoup(html, "html.parser")

        for tag in soup.findAll('div', class_="news-title"):
            for href in tag.findAll('a', href= True):
                urls.append(href.get_text())

    return urls
```

```python
headlines = allsides_headlines(urls)
len(headlines)
```

```python
# scrape sources from all sides

def allsides_sources(sites):
    
    sources = []

    for site in sites:

        url1 = f'https://www.allsides.com/{site}'


        response = requests.get(url1)
        html = response.content

        soup = BeautifulSoup(html, "html.parser")

        for tag in soup.findAll('div', class_="election-content container"):
            for item in tag.findAll('div', class_='news-source'):
                sources.append(item.get_text().strip())

    return sources
```

```python
sources = allsides_sources(urls)
len(sources)
```

```python
all_sides = pd.DataFrame({
    'headline': headlines,
    'source': sources,
    'url': urls_lst
})

all_sides
```

```python
# gets the text of an article from all the scraped urls

def get_text(urls, tag1):
    
    text = []

    for url in urls:

        response = requests.get(url)
        html = response.content

        soup = BeautifulSoup(html, "html.parser")

        if soup.find_all("div", class_= tag1):
            for tag in soup.findAll('div', class_= tag1):
                text.append(tag.get_text().strip())
        else:
            text.append("couldn't find text")

    return text
```

```python
text = get_text(urls_lst)
len(text)
```

```python
all_sides['text'] = text
all_sides
```

```python
all_sides.to_csv('all_sides.csv')
```

```python
# pandas df of all the scraped articles from all sides

as_df = pd.read_csv('all_sides.csv')
as_df
```

```python
# some text is missing because it has a different html tag

missing_text_df = as_df[as_df['text'].str.contains('find text')]
missing_text_df
```

```python
missing_text_urls = missing_text_df['url'].to_list()
len(missing_text_urls)
```

```python
# get the text for the missing rows

missing_text = get_text(missing_text_urls, 'story-id-page-description')
len(missing_text)
```

```python
# new df for the 115 entries which had missing text

replaced_text_df = missing_text_df.drop(columns= ['Unnamed: 0', 'text'])

# add col with the missing text scraped from all sides
replaced_text_df['text'] = missing_text

replaced_text_df = replaced_text_df.reset_index(drop= True)

replaced_text_df
```

```python
# drop rows from main df where text is missing

as_df_copy = as_df[as_df.text != 'couldn\'t find text']
as_df_copy = as_df_copy.reset_index(drop=True)

as_df_copy
```

```python
as_df_final = pd.concat([as_df_copy, replaced_text_df], ignore_index=True).drop(columns='Unnamed: 0')
as_df_final
```

```python
as_df_final.to_csv('as_df_final.csv')
```

```python
urls2 = ['https://apnews.com/hub/coronavirus-pandemic?utm_source=apnewsnav&utm_medium=featured']
```

```python
# get headlines of news articles from assocciated press

def ap_headlines(urls):

        headlines = []

        for url in urls:

                driver = webdriver.Chrome(ChromeDriverManager().install())

                driver.get(url)

                html = driver.find_elements_by_xpath("//*[@id='root']/div/main/div[3]/div/article/div[*]/div/a/h3")

                for element in range(len(html)):
                        try:
                                headlines.append(html[element].text)
                        except:
                                headlines.append('no headline')
        
                driver.close()

        return headlines
```

```python
headlines = ap_headlines(urls2)
len(headlines)
```

```python
# get the urls for each headline

def headline_urls_ap(urls):

    links = []

    for url in urls:

        driver = webdriver.Chrome(ChromeDriverManager().install())

        driver.get(url)

        html = driver.find_elements_by_xpath("//*[@id='root']/div/main/div[3]/div/article/div[*]/div/a")


        for element in html:
            try:
                links.append(element.get_attribute('href'))
            except:
                links.append('no url')
        
        driver.close()

    return links
```

```python
links = headline_urls_ap(urls1)
len(links)
```

```python
# get the text from the url

def text_from_url(urls):

    text_list = []

    for url in urls:

        response = requests.get(url)
        html = response.content

        soup = BeautifulSoup(html, "html.parser")

        try:
            text_list.append(soup.find("div", class_= 'Article').get_text())
        except:
            text_list.append('no text')
    
    return text_list
```

```python
text1 = text_from_url(links)
len(text1)
```

```python
# put everything in a pandas df

ap_df = pd.DataFrame(
    {'headline': headlines,
    'source': 'AP',
    'urls': links, 
    'text': text1,
    }
)
```

```python
ap_df.to_csv('ap_df.csv')
```

```python
# get the ending of each url from reuters
# we can use this in the next function to open each url and get the text and headline

def reuters_urls(urls):

    url_endings = []

    for url in urls:

        response = requests.get(url)
        html = response.content

        soup = BeautifulSoup(html, "html.parser")


        for div in soup.find_all('div', class_='StoryCollection__story___3EY8PG'):
            for href in div.find_all('a', href=True):
                url_endings.append(href['href'])

    return url_endings
```

```python
url_endings = reuters_urls(urls3)
len(url_endings)
```

```python
# open all the urls from above and gets the text and headline and saves it to a pandas df

def reuters_dataframe(url_ends):

    headlines = []
    text = []
    urls = []

    for end in url_ends:
        url = f'https://www.reuters.com{end}'
        urls.append(url)

        response = requests.get(url)
        html = response.content

        soup = BeautifulSoup(html, "html.parser")

        text.append(soup.find("p", class_="Text__text___3eVx1j Text__dark-grey___AS2I_p Text__regular___Bh17t- Text__large___1i0u1F Body__base___25kqPt Body__large_body___3g04wK ArticleBody__element___3UrnEs").get_text())
        
        headlines.append(soup.find("h1", class_ = "Text__text___3eVx1j Text__dark-grey___AS2I_p Text__medium___1ocDap Text__heading_2___sUlNJP Heading__base___1dDlXY Heading__heading_2___3f_bIW").get_text())


    df = pd.DataFrame(
        {'headline': headlines,
        'source': 'reuters',
        'urls': urls,
        'text': text
        })

    return df
```

```python
reuters_df = reuters_dataframe(url_endings)
reuters_df
```

```python
reuters_df.to_csv('reuters_df.csv')
```
